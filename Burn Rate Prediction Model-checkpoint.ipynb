```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
R = 8.314  # Gas constant (J/molÂ·K)
T0 = 300   # Initial temperature (K)
T_ign = 1200  # Ignition temperature (K)
L = 0.01   # Sample length (m)
dx = 0.00005  # Spatial step (m)
dt = 2e-7     # Smaller time step (s)
N = int(L / dx) + 1  # Number of spatial points
t_max = 0.005  # Max simulation time (s)

# Database of thermite compounds (refined parameters)
thermites = {
    "Al/CuO": {
        "rho": 3500,  # kg/mÂ³
        "Cp": 900,    # J/kgÂ·K
        "k_therm": 15,  # Higher thermal conductivity (W/mÂ·K)
        "Q": 4.5e6,   # Higher enthalpy (J/kg)
        "A": 5e12,    # Adjusted reaction rate (1/s)
        "Ea": 85e3,   # Lower activation energy (J/mol)
        "D0": 1e-5,   # Adjusted diffusion (mÂ²/s)
        "Ex": 65e3,   # Lower diffusion barrier (J/mol)
        "expected_burn_rate": 60  # cm/s
    },
    "Al/Fe2O3": {
        "rho": 3400,
        "Cp": 850,
        "k_therm": 10,
        "Q": 3.9e6,
        "A": 3e12,
        "Ea": 100e3,
        "D0": 8e-6,
        "Ex": 85e3,
        "expected_burn_rate": 41
    },
    "Al/WO3": {
        "rho": 3600,
        "Cp": 900,
        "k_therm": 7,
        "Q": 3.6e6,
        "A": 2e12,
        "Ea": 115e3,
        "D0": 6e-6,
        "Ex": 95e3,
        "expected_burn_rate": 35
    },
    "Al/Pb3O4": {
        "rho": 3700,
        "Cp": 900,
        "k_therm": 6,
        "Q": 3.5e6,
        "A": 1.5e12,
        "Ea": 120e3,
        "D0": 5e-6,
        "Ex": 100e3,
        "expected_burn_rate": 30
    },
    "Zr/Fe2O3": {
        "rho": 3800,
        "Cp": 800,
        "k_therm": 5,
        "Q": 3.2e6,
        "A": 1e12,
        "Ea": 140e3,
        "D0": 5e-6,
        "Ex": 120e3,
        "expected_burn_rate": 32
    },
    "B/Fe2O3": {
        "rho": 3000,
        "Cp": 1000,
        "k_therm": 4,
        "Q": 3.0e6,
        "A": 1e12,
        "Ea": 150e3,
        "D0": 3e-6,
        "Ex": 140e3,
        "expected_burn_rate": 16
    }
}

# Function to run simulation for a given thermite
def run_simulation(params):
    rho = params["rho"]
    Cp = params["Cp"]
    k_therm = params["k_therm"]
    Q = params["Q"]
    A = params["A"]
    Ea = params["Ea"]
    D0 = params["D0"]
    Ex = params["Ex"]
    
    alpha = k_therm / (rho * Cp)
    alpha_dt_dx2 = alpha * dt / (dx**2)
    D_dt_dx2 = D0 * dt / (dx**2)
    
    x = np.linspace(0, L, N)
    T = np.ones(N) * T0
    T[x < 0.0003] = 1500  # Smaller ignition pulse
    C = np.ones(N) * 1.0
    reactant = np.ones(N) * 1.0
    
    front_positions = []
    times = []
    max_reaction_rate = 0
    t = 0
    
    while t < t_max:
        T_new = T.copy()
        C_new = C.copy()
        reactant_new = reactant.copy()
        
        for i in range(1, N-1):
            reaction_rate = A * np.exp(-Ea / (R * T[i])) * reactant[i] * C[i]
            max_reaction_rate = max(max_reaction_rate, reaction_rate)
            Q_dot = Q * reaction_rate
            T_new[i] = T[i] + alpha_dt_dx2 * (T[i+1] - 2*T[i] + T[i-1]) + dt * Q_dot / (rho * Cp)
            
            D = D0 * np.exp(-Ex / (R * T[i]))
            C_new[i] = C[i] + D_dt_dx2 * (C[i+1] - 2*C[i] + C[i-1])
            
            reactant_new[i] = reactant[i] - dt * reaction_rate
        
        T_new[0] = T_new[-1] = T0
        C_new[0] = C_new[-1] = 1.0
        T = T_new
        C = C_new
        reactant = np.maximum(reactant_new, 0)
        
        front_idx = np.where(T > T_ign)[0]
        if len(front_idx) > 0:
            # Average front position over first few points to reduce noise
            front_pos = x[front_idx[:min(3, len(front_idx))]].mean()
            front_positions.append(front_pos)
            times.append(t)
        
        t += dt
    
    burning_rate = (front_positions[-1] - front_positions[0]) / (times[-1] - times[0]) * 100 if len(front_positions) > 1 else 0
    init_delay = times[0] if times else t_max
    return burning_rate, init_delay * 1e6, T, np.max(T), len(front_positions), max_reaction_rate

# Run simulations for all thermites
results = []
temperature_profiles = {}
for compound, params in thermites.items():
    burn_rate, init_delay, T, max_T, num_front_points, max_rate = run_simulation(params)
    results.append([compound, burn_rate, params["expected_burn_rate"], init_delay, max_T, num_front_points, max_rate])
    temperature_profiles[compound] = T

# Print results table
print("\nResults:")
print("-" * 100)
print(f"{'Compound':<12} {'Sim. Burn Rate (cm/s)':<20} {'Exp. Burn Rate (cm/s)':<20} {'Init. Delay (Âµs)':<15} {'Max Temp (K)':<15} {'Front Points':<15} {'Max Rate (1/s)':<15}")
print("-" * 100)
for res in results:
    print(f"{res[0]:<12} {res[1]:<20.2f} {res[2]:<20.2f} {res[3]:<15.2f} {res[4]:<15.2f} {res[5]:<15} {res[6]:<15.2e}")
print("-" * 100)

# Plot temperature profiles
plt.figure(figsize=(10, 6))
x = np.linspace(0, L, N)
for compound, T in temperature_profiles.items():
    plt.plot(x, T, label=compound)
plt.xlabel("Position (m)")
plt.ylabel("Temperature (K)")
plt.title("Temperature Profiles for Different Thermites")
plt.legend()
plt.show()
```
